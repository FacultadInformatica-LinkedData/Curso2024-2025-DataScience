Read the RDF file as shown in class

```python
from rdflib import Graph, Namespace, Literal
from rdflib.namespace import RDF, RDFS
g = Graph()
g.namespace_manager.bind('ns', Namespace("http://somewhere#"), override=False)
g.namespace_manager.bind('vcard', Namespace("http://www.w3.org/2001/vcard-rdf/3.0#"), override=False)
g.parse(github_storage+"/rdf/example5.rdf", format="xml")
```

Create a new class named Researcher

```python
ns = Namespace("http://somewhere#")
g.add((ns.Researcher, RDF.type, RDFS.Class))
for s, p, o in g:
  print(s,p,o)
```

**TASK 6.1: Create a new class named "University"**


```python
ns_1 = Namespace('http://somewhere#')
g.add((ns_1.University, RDF.type, RDFS.Class))
# Visualize the results
for s, p, o in g:
  print(s,p,o)
```

**TASK 6.2: Add "Researcher" as a subclass of "Person"**

```python
g.add((ns_1.Person, RDF.type, RDFS.Class))
g.add((ns_1.Researcher, RDFS.subClassOf, ns.Person))
# Visualize the results
for s, p, o in g:
  print(s,p,o)
```

**TASK 6.3: Create a new individual of Researcher named "Jane Smithers"**

```python
name = Literal('Jane Smithers')
uri_Jane = ns_1.JaneSmithers
g.add((uri_Jane, RDF.type, ns_1.Researcher))
# Visualize the results
for s, p, o in g:
  print(s,p,o)
```

**TASK 6.4: Add to the individual JaneSmithers the email address, fullName, given and family names. Use the https://schema.org vocabulary**

```python
sch = Namespace("https://schema.org/")
g.namespace_manager.bind('sch', Namespace("https://schema.org/"), override=False)
g.add((ns.JaneSmithers, sch.email, Literal("jsimthers@smwhere.com")))
for s, p, o in g:
  print(s,p,o)
```

**TASK 6.5: Add UPM as the university where John Smith works. Use the "https://example.org/ namespace**
